## Part E - The Company Locations

**Instructions:**
- **Determine** the company's unique locations (not including Remote).
- **Generate** the following output dynamically based on the provided data.
- **Must** sort locations alphabetically.

```python

# Your implementation:

unique_locations = sorted(set(locations) - {"Remote"})
num_of_locations = len(unique_locations)

print("{} has {} locations (not including remote):".format(company_name, num_of_locations))

for i, location in enumerate(unique_locations, start=1):
    print("{}. {}".format(i, location))
 
```

## Part F - The Remote Employees

**Instructions:**
- **Generate** the following output dynamically based on the provided data.
- **Must** display "employee" dynamically if only one (1) or "employees" if zero (0) or more than one (1).
- **Must** sort employees alphabetically by first name.

```python
# Your implementation:
remote_employees = [employee for employee, location in zip(employees, locations) if location == "Remote"]
num_remote_employees = len(remote_employees)

print("There are {} {} who work remotely at {}:".format(num_remote_employees, "employee" if num_remote_employees == 1 else "employees", company_name))

for i, employee in enumerate(sorted(remote_employees), start=1):
    print("{}. {}".format(i, employee))
```

## Part G - The Pricing Table
